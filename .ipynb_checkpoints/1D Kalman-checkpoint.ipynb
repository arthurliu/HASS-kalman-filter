```python
from random import Random
import matplotlib.pyplot as plt
%matplotlib inline
from collections import namedtuple
gaussian = namedtuple('Gaussian', ['mean', 'error']) # var

def predict(posterior, process):
    """Predict from posterior and process model."""
    posterior_mean, posterior_error = posterior
    process_mean, process_error = process
    predicted_state = posterior_mean + process_mean
    predicted_error = posterior_error + process_error
    return gaussian(predicted_state, predicted_error)

def update(prior, measurement):
    """Calc posterior from prior and measurement."""
    prior_mean, prior_error = prior
    measurement_mean, measurement_error = measurement
    
    residual = measurement_mean - prior_mean
    kalman_gain = prior_error / (prior_error + measurement_error)

    posterior_mean = prior_mean + kalman_gain*residual
    posterior_error = (1 - kalman_gain) * prior_error
    return gaussian(posterior_mean, posterior_error)
```

```python
random = Random()
random.gauss(mu=0, sigma=1)
```

```python
randoms = []
for i in range(1000000):
    randoms.append(random.normalvariate(mu=0, sigma=1))
plt.hist(randoms, bins=300);
```

The process variance (`process_var`) tells the filter how much the system is changing over time. If you set this number too low the filter will not be able to react to changes that are happening

```python
def volt(voltage, std):
    return voltage + (random.gauss(mu=0, sigma=1) * std)

voltage_std = .13  # Need to configure/estimate ORIG
process_var = .05**2  # Need to configure/estimate ORIG
#process_var = .5**2  # Need to configure/estimate HIGH
actual_voltage = 16.3

x = gaussian(25., 1000.) # initial state
process_model = gaussian(0., process_var)

N = 50
time = [i for i in range(N)]
readings = [volt(actual_voltage, voltage_std) for i in range(N)]
ps = []
estimates = []
```

```python
for z in readings:
    prior = predict(x, process_model)
    x = update(prior, gaussian(z, voltage_std**2))

    # save for latter plotting
    estimates.append(x.mean)
    ps.append(x.error)
```

```python
plt.figure(figsize=(12,5))
plt.plot(time, readings, 'bo', label='readings') # Dummy reading
plt.plot(time, estimates, 'ro--', label='estimates')
plt.legend()
#plt.ylim(16, 17)
plt.xlabel('step')
plt.ylabel('volts')
plt.show()
    
#plt.plot(ps)
#plt.title('Variance')
#print('Variance converges to {:.3f}'.format(ps[-1]))
```

Not bad, but note that our process model is a gaussian of mean zero. If there were drift, we would need to include this in the process model.

## Kalman 1D class
Lets create a simple class that we could use in Home-Assistant

```python
"""
Class for a 1D Kalman filter. The user will have to configure the filter sensitivity (range 0-1),
and the measurement standard deviation - if the process is stationary this is easy to calculate using 
one of the stats sensors.
"""

class kalman:
    """Class for a 1D Kalman filter."""
    
    def __init__(self, sensitivity):
        """Initialise the class"""
        self._gaussian = namedtuple('Gaussian', ['mean', 'var'])
        self._process_model = self._gaussian(0., (2*sensitivity)**2) # sensitivity should be in range 0-1
        self._prior = None
        self._state = None
        self._measurement_error = None
    
    def process_measurement(self, measurement):
        """Process a measurement"""
        if self._state is None and measurement is not None:
            self._measurement_error = 0.1 * measurement # Just guess variance to be 10%
            self._state = self._gaussian(measurement, self._measurement_error) # initial state and variance
        else:
            self._prior = self._predict(self._state, self._process_model)
            self._state = self._update(self._prior, self._gaussian(measurement, self._measurement_error))

    def _predict(self, posterior, process_model):
        """Predict from posterior and process model."""
        posterior_mean, posterior_error = posterior
        process_mean, process_error = process_model
        prior_state = posterior_mean + process_mean
        prior_error = posterior_error + process_error
        return self._gaussian(prior_state, prior_error)

    def _update(self, prior, measurement):
        """Calculate posterior from prior and measurement."""
        prior_mean, prior_error = prior
        measurement_mean, measurement_error = measurement
        residual = measurement_mean - prior_mean
        kalman_gain = prior_error / (prior_error + measurement_error)
        posterior_mean = prior_mean + kalman_gain*residual
        posterior_error = (1 - kalman_gain) * prior_error
        return self._gaussian(posterior_mean, posterior_error)

    @property
    def state(self):
        """Return the state of the entity."""
        return self._state.mean
    
    @property
    def variance(self):
        """Return the state of the entity."""
        return self._state.var
```

```python
sensitivity = 0.8 # range 0 - 1

kalman_1 = kalman(sensitivity)

states = []
variance = []
for z in readings:
    kalman_1.process_measurement(z)
    # save for latter plotting
    states.append(kalman_1.state)
    variance.append(kalman_1.variance)
```

```python
plt.figure(figsize=(17,7))
plt.plot(time, readings, 'bo', label='readings') # Dummy reading
plt.plot(time, states, 'ro--', label='states')
plt.plot(time, variance, 'go--', label='variance')
plt.legend()
#plt.ylim(16, 17)
plt.xlabel('step')
plt.ylabel('volts')
plt.show()
```
